```python
# cv2 implementation
import cv2
import math
import numpy as np
from google.colab.patches import cv2_imshow
```

### Task 1

```python
# resizing to 256 x 256
test_image = cv2.imread('coins.png')
new_dimentions = (256,256)
test_image_resized = cv2.resize(test_image, new_dimentions)
cv2_imshow(test_image_resized)
```

### Task 2

```python
# RGB to Gray Scale conversion
test_image_gray_scale = cv2.cvtColor(test_image_resized, cv2.COLOR_BGR2GRAY)
cv2_imshow(test_image_gray_scale)
```

### Task 3

```python
# RGB to Binary conversion via gray scale
test_image_binary = cv2.cvtColor(test_image_resized, cv2.COLOR_BGR2GRAY)
_, test_image_binary = cv2.threshold(test_image_binary, 157, 255, 0)
cv2_imshow(test_image_binary)
```

### Task 4

```python
gray_image = cv2.cvtColor(test_image_resized, cv2.COLOR_BGR2GRAY)

# applying GaussianBlur for removing white shine on coins
blur_image = cv2.GaussianBlur(gray_image,(11,11),cv2.BORDER_DEFAULT)

_, threshold_image = cv2.threshold(blur_image, 157, 255, cv2.THRESH_BINARY_INV + cv2.THRESH_OTSU)

# finding contours ( blob detection )
contours, _ = cv2.findContours(threshold_image, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

cv2_imshow(threshold_image)

print()
print("There are " + str(len(contours)) + " coins.")
```
